```python
import numpy as np
import matplotlib.pyplot as plt
from fFunc import Ffunc
from bak.fFunc import Ffunc as Ffunc_old
```

```python
f_func_obj = Ffunc(0.29, 0., 0.29)
f_func_obj_old = Ffunc_old(0.29, 0., 0.29)
```

```python
rs_exact = np.arange(0., 5.0, 0.1)
f_exacts = np.zeros(len(rs_exact))
f_exacts_old = np.zeros(len(rs_exact))
for i, r_exact in enumerate(rs_exact):
    f_exacts[i] = f_func_obj.f_exact(r_exact)
    f_exacts_old[i] = f_func_obj_old.f_exact(r_exact)
plt.plot(rs_exact, f_exacts, 'bo')
plt.plot(rs_exact, f_exacts_old, 'bx')

rs_interp = np.arange(0., 5.0, 0.01)
plt.plot(rs_interp, f_func_obj.f_interp(rs_interp), 'r-')
plt.plot(rs_interp, f_func_obj_old.f_interp(rs_interp), 'r--')
plt.show()
```
