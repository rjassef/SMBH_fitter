```python
import numpy as np
from fFunc import Ffunc
import matplotlib.pyplot as plt
```

```python
func_obj = Ffunc(0.1, 0., 0.1)
```

```python
rs_exact = np.arange(0., 5.0, 0.1)
rs_interp = np.arange(0., 5.0, 0.01)
plt.plot(rs_exact, func_obj.f_exact(rs_exact), 'bo')
plt.plot(rs_interp, func_obj.f_interp(rs_interp), 'r-')
plt.show()
```

```python
func_obj.f_interp(np.array([0.5, 0.7, 1.0, 0.9]))
```

```python
from gFunc import Gfunc
```

```python
gfunc_obj = Gfunc()
```

```python
n = 0.5
rnorms_exact = np.arange(0., 5.0, 0.25)
g_exact = np.zeros(len(rnorms_exact))
for i, rnorm in enumerate(rnorms_exact):
    g_exact[i] = gfunc_obj.g_exact(n, rnorm)
plt.plot(rnorms_exact, g_exact, 'bo')

rnorms_interp = np.arange(0., 5.0, 0.05)
plt.plot(rnorms_interp, gfunc_obj.g_interp((n, rnorms_interp)), 'r-')
```
